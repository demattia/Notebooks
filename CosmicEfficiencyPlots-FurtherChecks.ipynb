```python
def draw(name, normalize=False, yrange = 0, generator = False, ytitle = "", xtitle = ""):

    hData = inputFileData.Get(name)
    hData.SetTitle("")
    hMC = inputFileMC.Get(name)
    hMC.SetTitle("")
    hMC.SetLineColor(2)
    hMC.SetMarkerStyle(20)
    hMC.SetMarkerColor(2)
    hMC.SetMarkerSize(0.7)
    if generator:
        hMCGen = inputFileMC.Get("effVsGen"+name.split("effVs")[1])
        hMCGen.SetLineColor(4)
        hMCGen.SetMarkerStyle(22)
        hMCGen.SetMarkerSize(0.7)
        hMCGen.SetMarkerColor(4)

    if normalize:
        hData.Scale(1./hData.Integral())
        hMC.Scale(1./hMC.Integral())

    maxPlusErrorData = hData.GetBinContent(hData.GetMaximumBin())
    maxPlusErrorMC = hMC.GetBinContent(hMC.GetMaximumBin())
    # Because SetRangeUser and SetLimits do not work
    if maxPlusErrorData > maxPlusErrorMC:
        hData.Draw()
        hData.GetXaxis().SetTitle(xtitle)
        hData.GetXaxis().SetTitleOffset(0.95)
        hData.GetYaxis().SetTitle(ytitle)
        hData.GetYaxis().SetTitleOffset(0.95)
        if yrange != 0: hData.GetYaxis().SetRangeUser(0, yrange)
        hMC.Draw("same")
    else:
        hMC.Draw()
        hMC.GetXaxis().SetTitle(xtitle)
        hMC.GetXaxis().SetTitleOffset(0.95)
        hMC.GetYaxis().SetTitle(ytitle)
        hMC.GetYaxis().SetTitleOffset(0.95)
        if yrange != 0: hMC.GetYaxis().SetRangeUser(0, yrange)
        hData.Draw("same")
    if generator:
        hMCGen.Draw("same")
    # hData.GetYaxis().SetLimits(0, max(maxPlusErrorMC, maxPlusErrorData))

    leg = ROOT.TLegend(0.15,0.7,0.4,0.85)
    if generator:
        leg = ROOT.TLegend(0.15,0.15,0.4,0.4)
    leg.AddEntry(hData,"Data","ple")
    leg.AddEntry(hMC,"MC","ple")
    if generator:
        leg.AddEntry(hMCGen, "MC-truth","ple")
    leg.SetBorderSize(0)
    leg.SetFillColor(0)
    leg.Draw();

    return leg
```

Normalize the pt distribution from SA muons in $d_0$ bins to the same distribution from RSA muons. The RSA muons do not show a significant $d_0$ dependence and the efficiency is found to be in agreement between data and MC. If the SA muons $p_T$ distributions are also behaved similarly then the $p_T$ bias is well modeled by the simulation.

```python
inputFileData_RSA = ROOT.TFile("/home/demattia/cmslpc/uscms_data/d3/demattia/DisplacedLeptons/RefittedStandAloneMuons/Analysis/CMSSW_5_3_7/src/Analysis/TrackingEfficiencyFromCosmics/test/Macros/TreeAnalyzers/eff_data_dz50_RSA.root")
inputFileMC_RSA = ROOT.TFile("/home/demattia/cmslpc/uscms_data/d3/demattia/DisplacedLeptons/RefittedStandAloneMuons/Analysis/CMSSW_5_3_7/src/Analysis/TrackingEfficiencyFromCosmics/test/Macros/TreeAnalyzers/eff_MC_dz50_RSA.root")

def compute_ratio(h_SA, h_RSA, index, hType, rebin = 1):
    hRatio = h_SA.Clone("h"+str(index)+"Ratio"+hType)
    h_RSA_clone = h_RSA.Clone("h"+str(index)+"_RSA"+hType)
    hRatio.Sumw2()
    hRatio.Rebin(rebin)
    h_RSA_clone.Rebin(rebin)
    hRatio.Divide(hRatio, h_RSA_clone, 1/hRatio.GetEntries(), 1/h_RSA_clone.GetEntries())
    return hRatio

# MC
hPtBottom_recoD0_0_20_RSA = inputFileMC_RSA.Get("PtBottom_recoD0_0_20")
hPtBottom_recoD0_20_50_RSA = inputFileMC_RSA.Get("PtBottom_recoD0_20_50")
hPtBottom_recoD0_50_100_RSA = inputFileMC_RSA.Get("PtBottom_recoD0_50_100")
hPtBottom_recoD0_100_200_RSA = inputFileMC_RSA.Get("PtBottom_recoD0_100_200")
hPtBottom_recoD0_200_300_RSA = inputFileMC_RSA.Get("PtBottom_recoD0_200_300")
hPtBottom_recoD0_300_RSA = inputFileMC_RSA.Get("PtBottom_recoD0_300")
h0Ratio = compute_ratio(hPtBottom_recoD0_0_20, hPtBottom_recoD0_0_20_RSA, 0, "_MC", 4)
h1Ratio = compute_ratio(hPtBottom_recoD0_20_50, hPtBottom_recoD0_20_50_RSA, 1, "_MC", 4)
h2Ratio = compute_ratio(hPtBottom_recoD0_50_100, hPtBottom_recoD0_50_100_RSA, 2, "_MC", 4)
h3Ratio = compute_ratio(hPtBottom_recoD0_100_200, hPtBottom_recoD0_100_200_RSA, 3, "_MC", 4)
h4Ratio = compute_ratio(hPtBottom_recoD0_200_300, hPtBottom_recoD0_200_300_RSA, 4, "_MC", 4)
h5Ratio = compute_ratio(hPtBottom_recoD0_300, hPtBottom_recoD0_300_RSA, 5, "_MC", 4)

# Data
hPtBottom_recoD0_0_20_RSA_data = inputFileData_RSA.Get("PtBottom_recoD0_0_20")
hPtBottom_recoD0_20_50_RSA_data = inputFileData_RSA.Get("PtBottom_recoD0_20_50")
hPtBottom_recoD0_50_100_RSA_data = inputFileData_RSA.Get("PtBottom_recoD0_50_100")
hPtBottom_recoD0_100_200_RSA_data = inputFileData_RSA.Get("PtBottom_recoD0_100_200")
hPtBottom_recoD0_200_300_RSA_data = inputFileData_RSA.Get("PtBottom_recoD0_200_300")
hPtBottom_recoD0_300_RSA_data = inputFileData_RSA.Get("PtBottom_recoD0_300")
h0RatioData = compute_ratio(hPtBottom_recoD0_0_20_data, hPtBottom_recoD0_0_20_RSA_data, 0, "_Data", 4)
h1RatioData = compute_ratio(hPtBottom_recoD0_20_50_data, hPtBottom_recoD0_20_50_RSA_data, 1, "_Data", 4)
h2RatioData = compute_ratio(hPtBottom_recoD0_50_100_data, hPtBottom_recoD0_50_100_RSA_data, 2, "_Data", 4)
h3RatioData = compute_ratio(hPtBottom_recoD0_100_200_data, hPtBottom_recoD0_100_200_RSA_data, 3, "_Data", 4)
h4RatioData = compute_ratio(hPtBottom_recoD0_200_300_data, hPtBottom_recoD0_200_300_RSA_data, 4, "_Data", 4)
h5RatioData = compute_ratio(hPtBottom_recoD0_300_data, hPtBottom_recoD0_300_RSA_data, 5, "_Data", 4)

def save_canvas(hRatio, hRatioData, output_name, yRange, legendTitle):
    cRatio = TCanvas("cRatio", "cRatio", 400, 400)
    hRatioData.Draw("E")
    hRatioData.SetTitle("")
    hRatioData.GetYaxis().SetRangeUser(0, yRange)
    hRatioData.SetMarkerStyle(20)
    hRatioData.SetMarkerSize(0.8)
    hRatioData.GetXaxis().SetTitle("p^{SA}_{T} / p^{RSA}_{T}")
    hRatioData.GetXaxis().SetTitleOffset(0.95)
    hRatioData.GetYaxis().SetTitle("Arbitrary Units")
    hRatioData.GetYaxis().SetTitleOffset(0.95)
    hRatio.Draw("Esame")
    hRatio.SetMarkerStyle(21)
    hRatio.SetMarkerSize(0.8)
    hRatio.SetMarkerColor(2)
    hRatio.SetLineColor(2)
    leg = ROOT.TLegend(0.5, 0.7, 0.89, 0.89)
    leg.SetHeader(legendTitle)
    leg.SetFillColor(0)
    leg.SetBorderSize(0)
    leg.AddEntry(hRatioData, "Cosmics data", "lp")
    leg.AddEntry(hRatio, "Cosmics MC", "lp")
    leg.Draw("same")
    cRatio.SaveAs(output_name+".C")
    cRatio.SaveAs(output_name+".root")
    cRatio.SaveAs(output_name+".pdf")
    return leg


leg0 = save_canvas(h0Ratio, h0RatioData, "PtRatioSA_d0_0_20", 2.2, "|d_{0}| < 20 cm")
leg1 = save_canvas(h1Ratio, h1RatioData, "PtRatioSA_d0_20_50", 2.2, "20 < |d_{0}| < 50 cm")
leg2 = save_canvas(h2Ratio, h2RatioData, "PtRatioSA_d0_50_100", 2.5, "50 < |d_{0}| < 100 cm")
leg3 = save_canvas(h3Ratio, h3RatioData, "PtRatioSA_d0_100_200", 2.5, "100 < |d_{0}| < 200 cm")
leg4 = save_canvas(h4Ratio, h4RatioData, "PtRatioSA_d0_200_300", 2.5, "200 < |d_{0}| < 300 cm")
leg5 = save_canvas(h5Ratio, h5RatioData, "PtRatioSA_d0_300", 2.5, "|d_{0}| > 300 cm")


cRatio = TCanvas("cRatio", "cRatio", 800, 1200)
cRatio.Divide(2, 3)
cRatio.cd(1)
h0RatioData.Draw("E")
h0Ratio.Draw("Esame")
leg0.Draw("same")
cRatio.cd(2)
h1RatioData.Draw("E")
h1Ratio.Draw("Esame")
leg1.Draw("same")
cRatio.cd(3)
h2RatioData.Draw("E")
h2Ratio.Draw("Esame")
leg2.Draw("same")
cRatio.cd(4)
h3RatioData.Draw("E")
h3Ratio.Draw("Esame")
leg3.Draw("same")
cRatio.cd(5)
h4RatioData.Draw("E")
h4Ratio.Draw("Esame")
leg4.Draw("same")
cRatio.cd(6)
h5RatioData.Draw("E")
h5Ratio.Draw("Esame")
leg5.Draw("same")
cRatio

```
