Compute the efficiencies of the different selections with respect to the events within the acceptance.

```python
def drawEfficiency(denominator, y_title):
    eff_trigger = ROOT.TEfficiency(gen_lxy_all_DecayMode_withinAcc_trigFired, denominator)
    eff_reco = ROOT.TEfficiency(gen_lxy_reco_genMatched, denominator)
    eff_final = ROOT.TEfficiency(gen_lxy_final, denominator)
    c = ROOT.TCanvas()
    eff_trigger.Draw()
    eff_trigger.SetMarkerStyle(23)
    eff_trigger.SetMarkerSize(0.8)
    eff_reco.Draw("same")
    eff_final.Draw("same")
    eff_reco.SetLineColor(2)
    eff_reco.SetMarkerStyle(20)
    eff_reco.SetMarkerSize(0.5)
    eff_reco.SetMarkerColor(2)
    eff_final.SetLineColor(4)
    eff_final.SetMarkerStyle(22)
    eff_final.SetMarkerSize(0.8)
    eff_final.SetMarkerColor(4)
    eff_trigger.SetTitle("; generated L_{xy} [cm]; "+y_title)
    ROOT.gPad.Update()
    eff_trigger.GetPaintedGraph().GetYaxis().SetRangeUser(0, 1)

    # Legend
    leg = TLegend(0.2,0.7,0.85,0.9)
    leg.SetFillColor(0)
    leg.SetLineColor(0)
    leg.AddEntry(eff_trigger, "Trigger efficiency","lp")
    leg.AddEntry(eff_reco, "Reconstruction efficiency","lp")
    leg.AddEntry(eff_final, "Full selection efficiency","lp")
    leg.Draw("SAME")

    return c, eff_trigger, eff_reco, eff_final, leg
```

```python
c_full = drawEfficiency(gen_lxy_all_DecayMode, "Lxy-differential efficiency")
c_full[0].SaveAs("Eff_vs_Lxy.C")
c_full[0].SaveAs("Eff_vs_Lxy.root")
c_full[0].SaveAs("Eff_vs_Lxy.pdf")
c_full[0]
```

```python
c = drawEfficiency(gen_lxy_all_DecayMode_withinAcc, "Lxy-differential efficiency within acceptance")
c[0].SaveAs("EffWithinAcc_vs_Lxy.C")
c[0].SaveAs("EffWithinAcc_vs_Lxy.root")
c[0].SaveAs("EffWithinAcc_vs_Lxy.pdf")
c[0]
```

Acceptance is defined as
- $p_T > 21$ GeV/c
- $|\eta| < 2$
- gen-decay length < 500 cm
