```python
inputFileData = ROOT.TFile("/home/demattia/cmslpc/uscms_data/d3/demattia/DisplacedLeptons/RefittedStandAloneMuons/Analysis/CMSSW_5_3_7/src/Analysis/TrackingEfficiencyFromCosmics/test/Macros/TreeAnalyzers/TrackerTracks/eff_data_dz10.root")
inputFileMC = ROOT.TFile("/home/demattia/cmslpc/uscms_data/d3/demattia/DisplacedLeptons/RefittedStandAloneMuons/Analysis/CMSSW_5_3_7/src/Analysis/TrackingEfficiencyFromCosmics/test/Macros/TreeAnalyzers/TrackerTracks/eff_MC_dz10.root")
```

```python
def draw(name, normalize=False, yrange = 0):

    hData = inputFileData.Get(name)
    hMC = inputFileMC.Get(name)
    hMC.SetLineColor(2)

    if normalize:
        hData.Scale(1./hData.Integral())
        hMC.Scale(1./hMC.Integral())

    maxPlusErrorData = hData.GetBinContent(hData.GetMaximumBin())
    maxPlusErrorMC = hMC.GetBinContent(hMC.GetMaximumBin())
    # Because SetRangeUser and SetLimits do not work
    if maxPlusErrorData > maxPlusErrorMC:
        hData.Draw()
        if yrange != 0: hData.GetYaxis().SetRangeUser(0, yrange)
        hMC.Draw("same")
    else:
        hMC.Draw()
        if yrange != 0: hMC.GetYaxis().SetRangeUser(0, yrange)
        hData.Draw("same")
    # hData.GetYaxis().SetLimits(0, max(maxPlusErrorMC, maxPlusErrorData))

    leg = ROOT.TLegend(0.15,0.7,0.4,0.85);
    leg.AddEntry(hData,"Data");
    leg.AddEntry(hMC,"MC");
    leg.SetBorderSize(0)
    leg.Draw();

    return leg
```
