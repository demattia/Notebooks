```python
def draw(name, rebin):
    hData = inputFileData.Get(name)
    hMC = inputFileMC.Get(name)

    hData.Scale(1./hData.Integral())
    hMC.Scale(1./hMC.Integral())
    hData.RebinX(rebin)
    hMC.RebinX(rebin)

    hData.Draw()
    hMC.Draw("same")
    hMC.SetLineColor(2)

    leg = ROOT.TLegend(0.45,0.7,0.8,0.85);
    leg.AddEntry(hData,"Data");
    leg.AddEntry(hMC,"MC");
    leg.SetBorderSize(0)
    leg.Draw();
    
    return hData, hMC, leg
```

```python
c1 = TCanvas("c1", "c1", 800, 800)
l = draw("totalVsChi2_noWeight", 5)
c1.SetLogy(True)
c1
```

Do the reweighting by taking the ratio of the normalized histograms.

```python
c2 = TCanvas()
c2.SetLogy()
hWeight = l[0].Clone("TrackChi2Weight")
hWeight.Divide(l[1])
hWeight.Draw()
c2
```

```python
outputFile = ROOT.TFile("/home/demattia/cmslpc/uscms_data/d3/demattia/DisplacedLeptons/RefittedStandAloneMuons/Analysis/CMSSW_5_3_7/src/Analysis/TrackingEfficiencyFromCosmics/test/Macros/TreeAnalyzers/weight.root", "RECREATE")
hWeight.Write()
outputFile.Write()
outputFile.Close()
```

```python
def draw2D(name, xLabel, yLabel):
    c2 = TCanvas("c", "c", 1200, 2400)
    c2.Divide(1,2)
    c2.cd(1)
    hData = inputFileData.Get(name)
    hData.SetXTitle(xLabel)
    hData.SetYTitle(yLabel)
    hData.Draw("COLZ")
    c2.cd(2)
    hMC = inputFileMC.Get(name)
    hMC.SetXTitle(xLabel)
    hMC.SetYTitle(yLabel)
    hMC.Draw("COLZ")
    return c2
```

```python
c3 = TCanvas("c", "c", 1200, 2400)
c3.Divide(1,2)
c3.cd(1)
hDataHalf = inputFileData.Get("hEtaPhiHalf_noWeight")
hDataHalf.SetXTitle("#eta")
hDataHalf.SetYTitle("#phi")
hDataHalf.Draw("COLZ")
c3.cd(2)
hMCHalf = inputFileMC.Get("hEtaPhiHalf_noWeight")
hMCHalf.SetXTitle("#eta")
hMCHalf.SetYTitle("#phi")
hMCHalf.Draw("COLZ")
c3
```
