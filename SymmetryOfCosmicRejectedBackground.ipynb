# Symmetry of Cosmic-Rejected Background

We want to check the symmetry (or lack of) of the background rejected by the cosmic rejection cut. We run the analysis with and without the lifetime selection and save the cosmic-rejected events in output files. The files are read by this notebook and are printed.

More events are rejected in the control region before the lifetime (d0/sigma and Lxy/sigma) cuts. After the selection, more events are rejected in the signal region. The difference between the event numbers is less than two sigma (poission statistics).

```python
%pylab inline
```

```python
import matplotlib.pyplot as plt
from math import sqrt
import numpy
```

```python
def extract_events(file_name):
    input_file = open("/home/demattia/cmslpc//uscms_data/d3/demattia/DisplacedLeptons/RefittedStandAloneMuons/NoVertexFit/CMSSW_5_3_8/src/AnalysisMacros/Latest_Version_Code_June_19/"+file_name)
    found = False
    dPhiCorr = []
    for line in input_file:
        if found == True:
            if float(line.split("Lxy/sigma = ")[1].split(" ")[0]) > 9:
                dPhiCorr.append(float(line.split("dPhiCorr = ")[1]))
            found = False
        if line.find("Event") != -1:
            found = True
            # print line
    return dPhiCorr
        
```

```python
def plot_events(dPhiCorrControl, dPhiCorrSignal):
    bins = numpy.linspace(0, 3.15, 31)
    n1, bins1, patches1 = plt.hist(dPhiCorrControl+dPhiCorrSignal, bins, normed=0, facecolor='g', alpha=0.75, label='signal region')
    n2, bins2, patches2 = plt.hist(dPhiCorrControl, bins, normed=0, facecolor='b', alpha=0.75, label='control region')

    plt.xlabel('DeltaPhiCorr')
    plt.ylabel('Events')
    plt.title('Events rejected by the cosmic rejection cut')
    # plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
    # plt.axis([0, 3.15, 0, 0.03])
    plt.axis([0, 3.15, 0, 8])
    plt.grid(True)
    plt.legend(loc='upper left')

    plt.axvline(x=1.57, ymin=0, ymax=10, color='black', linestyle="--")

    plt.show()
```

```python
def results(control_name, signal_name):
    dPhiCorrControl = extract_events(control_name)
    dPhiCorrSignal = extract_events(signal_name)

    print "Control region:", len(dPhiCorrControl)
    print "Signal region:", len(dPhiCorrSignal)

    plot_events(dPhiCorrControl, dPhiCorrSignal)
```

```python
results("cosmicRejectedControlRegion.txt", "cosmicRejectedSignalRegion.txt")
```
