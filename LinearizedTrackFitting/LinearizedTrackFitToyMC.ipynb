The variables are the $\phi$ and $z$ coordinates of the hits.

```python
def get_plots(name):
    plots = []
    for i in range(0, maxRange):
        plots.append(input_file.Get(name+"_"+str(i)))
    return plots
```

```python
h_var = get_plots("Var")
h_PC = get_plots("PC")
h_PC_norm = get_plots("PCNorm")
```

```python
def draw_plots(name, plots, div_1 = 2, div_2 = 6, x_title_1 = "#phi", x_title_2 = "z"):
    c = TCanvas("c_"+name, "c_"+name, 800, 800)
    c.Divide(div_1, div_2)
    for i in range(0, maxRange):
        c.cd(i+1)
        plots[i].Draw()
        if i%2 == 0:
            plots[i].GetXaxis().SetTitle(x_title_1)
        else:
            plots[i].GetXaxis().SetTitle(x_title_2)
    return c
```

```python
c_var = draw_plots("var", h_var)
c_var
```

## Track Parameters

```python
def draw_params(prefix = "", fit = False):
    h_cot_theta = input_file.Get(prefix+"CotTheta")
    h_phi = input_file.Get(prefix+"Phi")
    h_z0 = input_file.Get(prefix+"Z0")
    h_inv_pt = input_file.Get(prefix+"InvPt")

    c_params = TCanvas(prefix+"cParams", prefix+"cParams", 800, 800)
    c_params.Divide(2, 2)
    c_params.cd(1)
    if fit:
        h_cot_theta.Fit("gaus")
    else:
        h_cot_theta.Draw()
    h_cot_theta.GetXaxis().SetTitle("cot(#theta)")
    c_params.cd(2)
    if fit:
        h_phi.Fit("gaus")
    else:
        h_phi.Draw()
    h_phi.GetXaxis().SetTitle("#phi")
    c_params.cd(3)
    if fit:
        h_z0.Fit("gaus")
    else:
        h_z0.Draw()
    h_z0.GetXaxis().SetTitle("z_{0}")
    c_params.cd(4)
    if fit:
        h_inv_pt.Fit("gaus")
    else:
        h_inv_pt.Draw()
    h_inv_pt.GetXaxis().SetTitle("1/p_{T}")

    return c_params, h_cot_theta, h_phi, h_z0, h_inv_pt
```
