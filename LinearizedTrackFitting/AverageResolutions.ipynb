```python
def fit(input_file_name, histo_name):
    input_file = TFile(input_file_name, "READ")
    h = input_file.FindObjectAny(histo_name)
    h.Fit("gaus")
    c = TCanvas("c", "c", 600, 600)
    h.Draw()
    mean = h.GetFunction("gaus").GetParameter(1)
    meanError = h.GetFunction("gaus").GetParError(1)
    sigma = h.GetFunction("gaus").GetParameter(2)
    sigmaError = h.GetFunction("gaus").GetParError(2)
    print "mean =", mean, "+/-", meanError
    print "sigma =", sigma, "+/-", sigmaError
    return c, h, input_file
```
