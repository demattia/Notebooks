```python
def draw(h1, h2, x_title, y_title):
    h1.Draw()
    h2.Draw("same")
    h1.GetXaxis().SetTitle(x_title)
    h1.GetXaxis().SetTitleOffset(1.3)
    h1.GetYaxis().SetTitle(y_title)
    h1.GetYaxis().SetTitleOffset(1.6)


def draw_superimposed(h1, h2, name, x_title, y_title="Entries"):
    if h1.GetMaximum() > h2.GetMaximum():
        draw(h1, h2, x_title, y_title)
    else:
        draw(h2, h1, x_title, y_title)
    h2.SetLineColor(2)
    l = TLegend(0.6,0.2,0.9,0.5)
    l.AddEntry(h1, name+"_{gen}")
    l.AddEntry(h2, name+"_{reco}")
    l.Draw("same")
    return l


def draw_and_fit(h, xTitle):
    h.Draw()
    h.Fit("gaus")
    # print h.GetFunction("gaus").GetParameter(0)
    mean = h.GetFunction("gaus").GetParameter(1)
    mean_error = h.GetFunction("gaus").GetParError(1)
    sigma = h.GetFunction("gaus").GetParameter(2)
    sigma_error = h.GetFunction("gaus").GetParError(2)
    dist = str(abs(int(math.log10(abs(mean_error))))+2)
    mean_string = ("{0:."+dist+"f}").format(mean) + " +/- " + ("{0:.2}").format(mean_error)
    sigma_string = ("{0:."+dist+"f}").format(sigma) + " +/- " + ("{0:.2}").format(sigma_error)
    h.GetXaxis().SetTitle(xTitle)
    h.GetXaxis().SetTitleOffset(1.3)
    h.GetYaxis().SetTitle("Entries")
    h.GetYaxis().SetTitleOffset(1.6)
    return mean_string, sigma_string


def plot_histograms(input_file_name, charge, par, par_name, unit):
    input_file = TFile(input_file_name, "READ")
    h_gen = input_file.FindObjectAny("pars_"+charge+"_"+par+";1")
    h_reco = input_file.FindObjectAny("EstimatedPar_"+charge+"_"+par+";1")
    h_error = input_file.FindObjectAny("EstimatedParError_"+charge+"_"+par+";1")
    h_relError = input_file.FindObjectAny("EstimatedParRelError_"+charge+"_"+par+";1")
    c = TCanvas("c", "c", 800, 800)
    c.Divide(2, 2)
    c.cd(1)
    l_1 = draw_superimposed(h_gen, h_reco, par_name, par_name+" "+unit)
    c.cd(2)
    e_m, e_s = draw_and_fit(h_error, par_name+"_{gen} - "+par_name+"_{reco} "+unit)
    c.cd(3)
    re_m, re_s = draw_and_fit(h_relError, "("+par_name+"_{gen} - "+par_name+"_{reco})/"+par_name+"_{gen}")
    return c, l_1, h_gen, h_reco, h_error, h_relError, input_file, e_m, e_s, re_m, re_s


def print_fit_result(c):
    print "absolute difference mean = "+c[7]
    print "absolute difference sigma = "+c[8]

    print "relative difference mean = "+c[9]
    print "relative difference sigma = "+c[10]
```

## Resolutions

```python
def compute_bins(pts, pt_min, pt_max, total_bins):
    bin_size = (pt_max - pt_min)/total_bins
    pt_bins = []
    for pt in pts:
        pt_bins.append(int((pt-pt_min)/bin_size)+1)
    return pt_bins


def fit_slices(input_file_name, h_name, bin_values, value_min, value_max, total_bins,
               mean_y_min, mean_y_max, sigma_y_min, sigma_y_max,
               x_title, mean_y_title, sigma_y_title):
    input_file = TFile(input_file_name, "READ")
    h = input_file.FindObjectAny(h_name)
    c1 = TCanvas()
    h.Draw("COLZ")
    c1

    bin_edges = compute_bins(bin_values, value_min, value_max, total_bins)
    print "bin_edges =", bin_edges
    projections = []
    bin_center = []
    bin_size = []
    mean = []
    meanError = []
    sigma = []
    sigmaError = []

    c2 = TCanvas("c2", "c2", 800, 1600)
    c2.Divide(2, len(bin_edges)/2)
    for i in range(len(bin_edges)-1):
        c2.cd(i+1)
        bin_center.append((bin_values[i+1]+bin_values[i])/2.)
        bin_size.append((bin_values[i+1]-bin_values[i])/2.)
        projY = h.ProjectionY(h.GetName()+"_py"+str(i), bin_edges[i], bin_edges[i+1]-1)
        if projY.GetEntries() > 0:
            projections.append(projY)
            projY.Fit("gaus")
            mean.append(projY.GetFunction("gaus").GetParameter(1))
            meanError.append(projY.GetFunction("gaus").GetParError(1))
            sigma.append(projY.GetFunction("gaus").GetParameter(2))
            sigmaError.append(projY.GetFunction("gaus").GetParError(2))
        else:
            mean.append(0.)
            meanError.append(0.)
            sigma.append(0.)
            sigmaError.append(0.)
            
    hMeans = TGraphErrors(len(bin_edges), array('d', bin_center), array('d', mean), array('d', bin_size), array('d', meanError))
    hSigmas = TGraphErrors(len(bin_edges), array('d', bin_center), array('d', sigma), array('d', bin_size), array('d', sigmaError))

    c3 = TCanvas("Resolutions", "Resolutions", 800, 400)
    c3.Divide(2, 1)
    c3.cd(1)
    hMeans.Draw("APE")
    hMeans.SetMarkerStyle(20)
    hMeans.SetMarkerSize(0.5)
    hMeans.GetYaxis().SetRangeUser(mean_y_min, mean_y_max)
    hMeans.GetXaxis().SetTitle(x_title)
    hMeans.GetYaxis().SetTitleOffset(1.5)
    hMeans.GetYaxis().SetTitle(mean_y_title)
    c3.cd(2)
    hSigmas.Draw("APE")
    hSigmas.SetMarkerStyle(20)
    hSigmas.SetMarkerSize(0.5)
    hSigmas.GetYaxis().SetRangeUser(sigma_y_min, sigma_y_max)
    hSigmas.GetXaxis().SetTitle(x_title)
    hSigmas.GetYaxis().SetTitleOffset(1.5)
    hSigmas.GetYaxis().SetTitle(sigma_y_title)
    c3
    
    return c1, c2, c3, hMeans, hSigmas, projections, h, bin_values, value_min, value_max, total_bins, input_file
```
