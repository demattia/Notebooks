```python
pylab.rcParams['figure.figsize'] = (8.0, 16.0)

x = np.linspace(-1.1*rho, 105, 10000) # 100 linearly spaced numbers
y = np.sqrt(-x**2+2*x*rho)

y_min = 102.

R6_shift_pos = 109.
R6_shift_neg = 106.5

# plt.axis([32.65, 34.65, y0, 105])


plt.axis([0, 40, 0, 120])



# plt.plot([30, 38], [105.5, 103], 'k-')

x6 = R6**2/(2*rho)
y6 = R6*sqrt(1-R6**2/(4*rho**2))

x6_shift_neg = R6_shift_neg**2/(2*rho)
y6_shift_neg = R6_shift_neg*sqrt(1-R6_shift_neg**2/(4*rho**2))

x6_shift_pos = R6_shift_pos**2/(2*rho)
y6_shift_pos = R6_shift_pos*sqrt(1-R6_shift_pos**2/(4*rho**2))

plt.plot([0, x6_shift_pos], [0, y6_shift_pos], 'k-')
plt.plot([0, x6], [0, y6], 'k-', linestyle="dashed")
plt.plot([0, x6_shift_neg], [0, y6_shift_neg], 'k-')

circle6 = plt.Circle((0,0), R6, facecolor='none', edgecolor="black", linestyle="dashed")
circle6_shift_neg = plt.Circle((0,0), R6_shift_neg, facecolor='none', edgecolor="black")
circle6_shift_pos = plt.Circle((0,0), R6_shift_pos, facecolor='none', edgecolor="black")
# wedge6 = mpatches.Wedge((0,0), 10, 0, 72, width = 1)
wedge6 = mpatches.Wedge((y_min*x6/y6, y_min), 0.1, 0, 72, width = 0.02)
wedge6_shift_neg = mpatches.Wedge((y_min*x6_shift_neg/y6_shift_neg, y_min), 0.1, 0, 72, width = 0.02)
wedge6_shift_pos = mpatches.Wedge((y_min*x6_shift_pos/y6_shift_pos, y_min), 0.1, 0, 72, width = 0.02)

fig = plt.gcf()
fig.gca().add_artist(circle6)
fig.gca().add_artist(circle6_shift_neg)
fig.gca().add_artist(circle6_shift_pos)
fig.gca().add_artist(wedge6)
fig.gca().add_artist(wedge6_shift_neg)
fig.gca().add_artist(wedge6_shift_pos)

plt.xlabel('x [cm]')
plt.ylabel('y [cm]')

pylab.plot(x, y)

plt.savefig("OutermostRadiusVariation.pdf")
```

```python
pylab.rcParams['figure.figsize'] = (8.0, 8.0)

x = np.linspace(-1.1*rho, 105, 10000) # 100 linearly spaced numbers
y = np.sqrt(-x**2+2*x*rho)

x_min = 31.65
x_max = 34.1

y_min = 100.
y_max = 105.

plt.axis([x_min, x_max, y_min, y_max])


# plt.axis([0, 40, 0, 120])


pylab.rcParams['figure.figsize'] = (8.0, 8.0*(y_min-y_max)/(x_min-x_max))



# plt.plot([30, 38], [105.5, 103], 'k-')

x6 = R6**2/(2*rho)
y6 = R6*sqrt(1-R6**2/(4*rho**2))

x6_shift_neg = R6_shift_neg**2/(2*rho)
y6_shift_neg = R6_shift_neg*sqrt(1-R6_shift_neg**2/(4*rho**2))

x6_shift_pos = R6_shift_pos**2/(2*rho)
y6_shift_pos = R6_shift_pos*sqrt(1-R6_shift_pos**2/(4*rho**2))

plt.plot([0, x6_shift_pos], [0, y6_shift_pos], 'k-')
plt.plot([0, x6], [0, y6], 'k-', linestyle="dashed")
plt.plot([0, x6_shift_neg], [0, y6_shift_neg], 'k-')

circle6 = plt.Circle((0,0), R6, facecolor='none', edgecolor="black", linestyle="dashed")
circle6_shift_neg = plt.Circle((0,0), R6_shift_neg, facecolor='none', edgecolor="black")
circle6_shift_pos = plt.Circle((0,0), R6_shift_pos, facecolor='none', edgecolor="black")
wedge6 = mpatches.Wedge((y_min*x6/y6, y_min), 0.1, 0, 72, width = 0.02)
wedge6_shift_neg = mpatches.Wedge((y_min*x6_shift_neg/y6_shift_neg, y_min), 0.1, 0, 72, width=0.02, facecolor="red")
wedge6_shift_pos = mpatches.Wedge((y_min*x6_shift_pos/y6_shift_pos, y_min), 0.1, 0, 72, width=0.02, facecolor="green")

fig = plt.gcf()
fig.gca().add_artist(circle6)
fig.gca().add_artist(circle6_shift_neg)
fig.gca().add_artist(circle6_shift_pos)
fig.gca().add_artist(wedge6)
fig.gca().add_artist(wedge6_shift_neg)
fig.gca().add_artist(wedge6_shift_pos)


plt.xlabel('x [cm]')
plt.ylabel('y [cm]')

phi_shift_neg = float(int(asin(x6_shift_neg/y6_shift_neg)*1000)/1000.)
phi_average = float(int(asin(x6/y6)*1000)/1000.)
phi_shift_pos = float(int(asin(x6_shift_pos/y6_shift_pos)*1000)/1000.)

plt.text(31.95, 100.2, r'$\phi = '+str(phi_shift_neg)+'$', fontsize=15)
plt.text(32.375, 100.2, r'$\phi = '+str(phi_average)+'$', fontsize=15)
plt.text(32.8, 100.2, r'$\phi = '+str(phi_shift_pos)+'$', fontsize=15)

pylab.plot(x, y)

plt.savefig("OutermostRadiusVariationZoom.pdf")
```

A roughly 2% effect that will reflect on the $p_T$ measurement. To bring down the $p_T$ resolution to the < 1% level this effect must be corrected. This effect appears in all layers and will lead to biases in the measured $p_T$ of similar magnitude if not corrected.

This effect is $p_T$ dependent: the shift of the $\phi$ coordinate is maximum for the lowest $p_T$ while it tends to zero as the $p_T$ tends to infinity (for tracks with null transverse impact parameter).

###Effect of the jumps in the radius on $p_T$

```python
# Innermost layer
phi_i_neg = asin(R1_shift_neg/(2*rho))
phi_i_pos = asin(R1_shift_pos/(2*rho))

# Compute the pt that would give the phi measured in the outermost module in the innermost module
pt_prime_pos_to_neg = R1_shift_neg/(2*sin(phi_i_pos))*3.8*0.003
# Compute the pt that would give the phi measured in the innermost module in the outermost module
pt_prime_neg_to_pos = R1_shift_pos/(2*sin(phi_i_neg))*3.8*0.003

print "Innermost: Percent change in phi =", (phi_i_pos-phi_i_neg)/phi_i_pos
print "Innermost: Percent change in pt (outer to inner) =", (pt_prime_pos_to_neg - pt)/pt
print "Innermost: Percent change in pt (outer to inner) =", (pt_prime_neg_to_pos - pt)/pt

# Outermost layer
phi_i_neg = asin(R6_shift_neg/(2*rho))
phi_i_pos = asin(R6_shift_pos/(2*rho))

# Compute the pt that would give the phi measured in the outermost module in the innermost module
pt_prime_pos_to_neg = R6_shift_neg/(2*sin(phi_i_pos))*3.8*0.003
# Compute the pt that would give the phi measured in the innermost module in the outermost module
pt_prime_neg_to_pos = R6_shift_pos/(2*sin(phi_i_neg))*3.8*0.003

print "Outermost: Percent change in phi =", (phi_i_pos-phi_i_neg)/phi_i_pos
print "Outermost: Percent change in pt (outer to inner) =", (pt_prime_pos_to_neg - pt)/pt
print "Outermost: Percent change in pt (outer to inner) =", (pt_prime_neg_to_pos - pt)/pt
```

###Effects of the non-flatness in the innermost layer on $p_T$

```python
phi_i_neg = asin(23/(2*rho))
phi_i_pos = asin(23.5/(2*rho))

# Compute the pt that would give the phi measured in the outermost module in the innermost module
pt_prime_pos_to_neg = 23/(2*sin(phi_i_pos))*3.8*0.003
pt_prime_neg_to_pos = 23.5/(2*sin(phi_i_neg))*3.8*0.003

print "Percent change in phi =", (phi_i_pos-phi_i_neg)/phi_i_pos
print "Percent change in pt (outer to inner) =", (pt_prime_pos_to_neg - pt)/pt
print "Percent change in pt (outer to inner) =", (pt_prime_neg_to_pos - pt)/pt
```

### Effects of the second order term in $z$

```python
second_order_z_term = (R1**3)/(rho**2)/24
print "Innermost: First order term =", R1
print "Innermost: Second order term =", second_order_z_term
print "Innermost: Relative variation =", second_order_z_term/R1

second_order_z_term = (R6**3)/(rho**2)/24
print "Outermost: First order term =", R6
print "Outermost: Second order term =", second_order_z_term
print "Outermost: Relative variation =", second_order_z_term/R6
```

### Effects of the jumps in the radius on $z$

```python
cotTheta = 0.8
z_0 = 0.
z1 = z_0 + 2*rho*cotTheta*asin(R1/(2*rho))
z1_neg = z_0 + 2*rho*cotTheta*asin(R1_shift_neg/(2*rho))
z1_pos = z_0 + 2*rho*cotTheta*asin(R1_shift_pos/(2*rho))

# Compute the cotTheta that would give the phi measured in the outermost module in the innermost module
cotThetaPrime_pos_to_neg = (z1_pos - z_0)/(2*rho*asin(R1_shift_neg/(2*rho)))
# Compute the cotTheta that would give the phi measured in the innermost module in the outermost module
cotThetaPrime_neg_to_pos = (z1_neg - z_0)/(2*rho*asin(R1_shift_pos/(2*rho)))
print "Innermost: Percent change in z =", (z1_neg - z1_pos)/z1_neg
print "Innermost: Percent change in cotTheta (outer to inner) =", (cotThetaPrime_pos_to_neg - cotTheta)/cotTheta
print "Innermost: Percent change in cotTheta (inner to outer) =", (cotThetaPrime_neg_to_pos - cotTheta)/cotTheta

z6 = z_0 + 2*rho*cotTheta*asin(R6/(2*rho))
z6_neg = z_0 + 2*rho*cotTheta*asin(R6_shift_neg/(2*rho))
z6_pos = z_0 + 2*rho*cotTheta*asin(R6_shift_pos/(2*rho))

# Compute the cotTheta that would give the phi measured in the outermost module in the innermost module
cotThetaPrime_pos_to_neg = (z6_pos - z_0)/(2*rho*asin(R6_shift_neg/(2*rho)))
# Compute the cotTheta that would give the phi measured in the innermost module in the outermost module
cotThetaPrime_neg_to_pos = (z6_neg - z_0)/(2*rho*asin(R6_shift_pos/(2*rho)))
print "Outermost: Percent change in z =", (z6_neg - z6_pos)/z6_neg
print "Outermost: Percent change in cotTheta (outer to inner) =", (cotThetaPrime_pos_to_neg - cotTheta)/cotTheta
print "Outermost: Percent change in cotTheta (inner to outer) =", (cotThetaPrime_neg_to_pos - cotTheta)/cotTheta
```
