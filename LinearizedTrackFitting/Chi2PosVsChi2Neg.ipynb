# Chi2 with Matrix for Negative Charge vs Chi2 with Matrix for Positive Charge

The matrices are built separately for muons and anti-muons with phi and R stub coordinates. The $\chi^2/ndof$ is evaluated from 12x12 covariance matrix of the stub coordinates using the 12 - 2 = 10 principal components associated to the smaller eigenvalues. In this notebook, when we mention $\chi^2$ we always refer to the normalized $\chi^2/ndof$.

```python
def draw_plot(h, option, range_x_min=0, range_x_max=0, range_y_min=0, range_y_max=0):
    h.Draw(option)
    if range_x_min != range_x_max:
        h.GetXaxis().SetRangeUser(range_x_min, range_x_max)
    if range_y_min != range_y_max:
        h.GetYaxis().SetRangeUser(range_y_min, range_y_max)
    h.GetXaxis().SetTitle("positive #chi2/ndof")
    h.GetXaxis().SetTitleOffset(1.3)
    h.GetYaxis().SetTitle("negative #chi2/ndof")
    h.GetYaxis().SetTitleOffset(1.3)
```

### Compute the fraction of tracks were the smallest chi2 would correspond to the incorrect charge.

We look at the value of the minimum #chi^2 between the positive and negative fit. The following histograms are filled only for the minimum of the two. Therefore, for instance, the plot of negative chi2 for positive charge muons contains only the chi2 values for positive charge muons computed with the negative fit when this is smaller than the chi2 from the positive fit.

```python
def draw_text(text):
    xlabel = TText()
    xlabel.SetNDC()
    xlabel.SetTextFont(1)
    xlabel.SetTextColor(1)
    xlabel.SetTextSize(0.03)
    xlabel.SetTextAlign(22)
    xlabel.SetTextAngle(0)
    xlabel.DrawText(0.5, 1., text)
    return xlabel
```

```python
def draw_chi2(h):
    h.Draw()
    h.GetXaxis().SetTitle("#chi2/ndof")
    h.GetXaxis().SetTitleOffset(1.3)
    h.GetYaxis().SetTitle("Entries")
    h.GetYaxis().SetTitleOffset(1.3)
```

```python
c_chi2 = TCanvas("c_chi2", "c_chi2", 800, 800)
c_chi2.Divide(2, 2)
c_chi2.cd(1)
draw_chi2(hChi2Pos_NegTk)
# draw_text("positive matrix on muons")
c_chi2.cd(2)
draw_chi2(hChi2Neg_NegTk)
# draw_text("negative matrix on muons")
c_chi2.cd(3)
draw_chi2(hChi2Pos_PosTk)
# draw_text("positive matrix on anti-muons")
c_chi2.cd(4)
draw_chi2(hChi2Neg_PosTk)
# draw_text("negative matrix on anti-muons")
c_chi2
```

```python
correct_neg = hChi2Neg_NegTk.GetEntries()
wrong_neg = hChi2Pos_NegTk.GetEntries()
correct_pos = hChi2Pos_PosTk.GetEntries()
wrong_pos = hChi2Neg_PosTk.GetEntries()

print "Number of muons with correct charge =", correct_neg
print "Number of muons with wrong charge =", wrong_neg
print "Number of anti-muons with correct charge =", correct_pos
print "Number of anti-muons with wrong charge =", wrong_pos
print
print "Fraction of muons with incorrect charge =", wrong_neg/(correct_neg+wrong_neg)
print "Percentage of correctly identified charge for muons =", correct_neg/(correct_neg+wrong_neg)*100., "%"
print
print "Fraction of anti-muons with incorrect charge =", wrong_pos/(correct_pos+wrong_pos)
print "Percentage of correctly identified charge for anti-muons =", correct_pos/(correct_pos+wrong_pos)*100., "%"
print
total_fraction_wrong = (wrong_pos+wrong_neg)/(correct_neg+correct_pos+wrong_neg+wrong_pos)
print "Total fraction of muons and anti-muons with incorrect charge =", total_fraction_wrong
total_percentage_correct = (correct_pos+correct_neg)/(correct_neg+correct_pos+wrong_neg+wrong_pos)*100.
print "Percentage of correctly identified charge for muons and anti-muons =", total_percentage_correct, "%"
```
