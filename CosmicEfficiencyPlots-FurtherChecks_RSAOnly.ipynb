```python
def draw(name, normalize=False, yrange = 0, generator = False, ytitle = "", xtitle = ""):

    hData = inputFileData.Get(name)
    hData.SetTitle("")
    hMC = inputFileMC.Get(name)
    hMC.SetTitle("")
    hMC.SetLineColor(2)
    hMC.SetMarkerStyle(20)
    hMC.SetMarkerColor(2)
    hMC.SetMarkerSize(0.7)
    if generator:
        hMCGen = inputFileMC.Get("effVsGen"+name.split("effVs")[1])
        hMCGen.SetLineColor(4)
        hMCGen.SetMarkerStyle(22)
        hMCGen.SetMarkerSize(0.7)
        hMCGen.SetMarkerColor(4)

    if normalize:
        hData.Scale(1./hData.Integral())
        hMC.Scale(1./hMC.Integral())

    maxPlusErrorData = hData.GetBinContent(hData.GetMaximumBin())
    maxPlusErrorMC = hMC.GetBinContent(hMC.GetMaximumBin())
    # Because SetRangeUser and SetLimits do not work
    if maxPlusErrorData > maxPlusErrorMC:
        hData.Draw()
        hData.GetXaxis().SetTitle(xtitle)
        hData.GetXaxis().SetTitleOffset(0.95)
        hData.GetYaxis().SetTitle(ytitle)
        hData.GetYaxis().SetTitleOffset(0.95)
        if yrange != 0: hData.GetYaxis().SetRangeUser(0, yrange)
        hMC.Draw("same")
    else:
        hMC.Draw()
        hMC.GetXaxis().SetTitle(xtitle)
        hMC.GetYaxis().SetTitle(ytitle)
        hMC.GetXaxis().SetTitleOffset(0.95)
        hMC.GetYaxis().SetTitleOffset(0.95)
        if yrange != 0: hMC.GetYaxis().SetRangeUser(0, yrange)
        hData.Draw("same")
    if generator:
        hMCGen.Draw("same")
    # hData.GetYaxis().SetLimits(0, max(maxPlusErrorMC, maxPlusErrorData))

    leg = ROOT.TLegend(0.15,0.7,0.4,0.85)
    if generator:
        leg = ROOT.TLegend(0.15,0.15,0.4,0.4)
    leg.AddEntry(hData,"Data","ple")
    leg.AddEntry(hMC,"MC","ple")
    if generator:
        leg.AddEntry(hMCGen, "MC-truth","ple")
    leg.SetBorderSize(0)
    leg.SetFillColor(0)
    leg.Draw();

    return leg
```
