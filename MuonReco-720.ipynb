# Reason for Low Trigger Efficiency

We explore the reason for the low trigger efficiency.

Note: this notebook requires to run the TriggerEfficiencyPlots.py FWLite macro on CMSSW root files. The collections required by the macro are listed therein and are StandAloneMuons, RefittedStandAloneMuons and L2Muons.

```python
def saveToPdf(h_name, x_title = "p_{T} [GeV/c]", y_title = "Entries", draw_options = ""):
    c = TCanvas(h_name+"_canvas", h_name+"_canvas", 400, 400)
    h = input_file.Get(h_name)
    # Ugly workaround to make the palette draw properly and accessible
    if draw_options == "COLZ":
        h.Draw()
        c.Update()
        h.Draw(draw_options)
        c.Update()
        h.Draw(draw_options)
    else:
        h.Draw(draw_options)
    h.SetTitle("")
    h.GetXaxis().SetTitle(x_title)
    h.GetXaxis().SetTitleOffset(1.4)
    h.GetXaxis().SetLabelSize(0.035)
    h.GetXaxis().SetTitleSize(0.035)
    h.GetYaxis().SetTitle(y_title)
    h.GetYaxis().SetTitleOffset(1.4)
    h.GetYaxis().SetTitleSize(0.035)
    h.GetYaxis().SetLabelSize(0.035)
    if draw_options == "COLZ":
        pal = h.GetListOfFunctions().FindObject("palette")
        if pal:
            pal.GetAxis().SetLabelSize(0.035)
            pal.GetAxis().SetTitleSize(0.035)
            # pal.GetAxis().SetTitleOffset(1.4)
            pal.GetAxis().SetTitle("Entries")
    c.Update()
    return h, c
```

```python
hc = dict()

name_list = []
name_list.append("genPtLow")
name_list.append("genPtHigh")
name_list.append("MuonPtLow")
name_list.append("MuonPtHigh")
name_list.append("OfflinePtLow")
name_list.append("OfflinePtHigh")
name_list.append("SAPtLow")
name_list.append("SAPtHigh")

for name in name_list:
    hc[name] = saveToPdf(name)
    hc[name+"_Lxy_0_50"] = saveToPdf(name+"_Lxy_0_50")
    hc[name+"_Lxy_50_100"] = saveToPdf(name+"_Lxy_50_100")
    hc[name+"_Lxy_100_200"] = saveToPdf(name+"_Lxy_100_200")
    hc[name+"_Lxy_200"] = saveToPdf(name+"_Lxy_200")

hc["genLxy"] = saveToPdf("genLxy", x_title = "L_{xy} [cm]")

hc["genLxyAndPt"] = saveToPdf("genLxyAndPt", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndD0"] = saveToPdf("genLxyAndD0", "gen L_{xy} [cm]", "d_{0} [cm]", "COLZ")
hc["genLxyAndGenPt"] = saveToPdf("genLxyAndGenPt", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndOfflinePt"] = saveToPdf("genLxyAndOfflinePt", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndOfflineD0"] = saveToPdf("genLxyAndOfflineD0", "gen L_{xy} [cm]", "d_{0} [cm]", "COLZ")
hc["genLxyAndSAPt"] = saveToPdf("genLxyAndSAPt", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndSAD0"] = saveToPdf("genLxyAndSAD0", "gen L_{xy} [cm]", "d_{0} [cm]", "COLZ")

hc["genLxyAndPtTriggered"] = saveToPdf("genLxyAndPtTriggered", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndD0Triggered"] = saveToPdf("genLxyAndD0Triggered", "gen L_{xy} [cm]", "d_{0} [cm]", "COLZ")
hc["genLxyAndGenPtTriggered"] = saveToPdf("genLxyAndGenPtTriggered", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndOfflinePtTriggered"] = saveToPdf("genLxyAndOfflinePtTriggered", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndOfflineD0Triggered"] = saveToPdf("genLxyAndOfflineD0Triggered", "gen L_{xy} [cm]", "d_{0} [cm]", "COLZ")
hc["genLxyAndSAPtTriggered"] = saveToPdf("genLxyAndSAPtTriggered", "gen L_{xy} [cm]", "p_{T} [GeV/c]", "COLZ")
hc["genLxyAndSAD0Triggered"] = saveToPdf("genLxyAndSAD0Triggered", "gen L_{xy} [cm]", "d_{0} [cm]", "COLZ")
```

```python
def pt_comparison(pt_type = "PtLow", leg_name_part = "Second"):
    c = TCanvas("", "", 400, 400)
    c.SetLogx(True)
    # c.SetLogy(True)
    # hc["MuonPtLow"][0].Draw()
    h1 = hc["Muon"+pt_type][0].Clone("Muon"+pt_type+"_clone")
    h1.Draw()
    h1.RebinX(2)
    h2 = hc["Offline"+pt_type][0].Clone("Offline"+pt_type+"_clone")
    h2.Draw("same")
    h2.RebinX(2)
    h2.SetLineColor(2)
    h3 = hc["SA"+pt_type][0].Clone("SA"+pt_type+"_clone")
    h3.Draw("same")
    h3.RebinX(2)
    h3.SetLineColor(4)
    h1.GetYaxis().SetRangeUser(0, 1.3*max(max(h1.GetMaximum(), h2.GetMaximum()), h3.GetMaximum()))
    leg = ROOT.TLegend(0.45,0.72,0.86,0.86)
    leg.SetFillColor(0)
    leg.SetBorderSize(0)
    leg.AddEntry(h1, leg_name_part+" p_{T} L2Muon")
    leg.AddEntry(h2, leg_name_part+" p_{T} RSA muon")
    leg.AddEntry(h3, leg_name_part+" p_{T} SA muon")
    leg.Draw("same")
    c.SaveAs(dir_name+"/"+pt_type+"Comparison.pdf")
    return c, h1, h2, h3, leg
```

```python
def draw_stack(h_name, option, rebinX = 0, normalize = False, draw_gen = False, logX = False, title = ""):
    c = TCanvas(h_name+"canvas_stack", h_name+"canvas_stack", 400, 400)
    if logX:
        c.SetLogx(True)
    # c.SetLogy(True)
    stack = ROOT.THStack("s", title+"; p_{T} [GeV/c]; Entries")
    h1 = hc[h_name+"_Lxy_0_50"][0].Clone(h_name+"_Lxy_0_50_stack")
    stack.Add(h1)
    h2 = hc[h_name+"_Lxy_50_100"][0].Clone(h_name+"_Lxy_50_100_stack")
    stack.Add(h2)
    h3 = hc[h_name+"_Lxy_100_200"][0].Clone(h_name+"_Lxy_100_200_stack")
    stack.Add(h3)
    h4 = hc[h_name+"_Lxy_200"][0].Clone(h_name+"_Lxy_200_stack")
    stack.Add(h4)
    legend_type = "f"
    if option == "nostack":
        legend_type = "l"
        h1.SetLineColor(2)
        h2.SetLineColor(3)
        h3.SetLineColor(4)
        h4.SetLineColor(1)
        h1.SetLineWidth(1)
        h2.SetLineWidth(1)
        h3.SetLineWidth(1)
        h4.SetLineWidth(1)
    else:
        h1.SetFillColor(2)
        h2.SetFillColor(3)
        h3.SetFillColor(4)
        h4.SetFillColor(5)
    if rebinX > 0:
        h1.RebinX(rebinX)
        h2.RebinX(rebinX)
        h3.RebinX(rebinX)
        h4.RebinX(rebinX)
    if normalize:
        h1.Scale(1/h1.GetEntries())
        h2.Scale(1/h2.GetEntries())
        h3.Scale(1/h3.GetEntries())
        h4.Scale(1/h4.GetEntries())
    stack.Draw(option)
    stack.GetHistogram().GetXaxis().SetTitleOffset(1.3);
    stack.GetHistogram().GetYaxis().SetTitleOffset(1.5);
    leg = ROOT.TLegend(0.45,0.65,0.85,0.85);
    leg.SetBorderSize(0)
    leg.SetFillColor(0)
    leg.AddEntry(h1,"p_{T} for 0 < |L_{xy}| < 50 cm",legend_type);
    leg.AddEntry(h2,"p_{T} for 50 < |L_{xy}| < 100 cm",legend_type);
    leg.AddEntry(h3,"p_{T} for 100 < |L_{xy}| < 200 cm",legend_type);
    leg.AddEntry(h4,"p_{T} for |L_{xy}| > 200 cm",legend_type);

    if draw_gen:
        gen_name = "genPtLow"
        if h_name.find("High") != -1:
            gen_name = "genPtHigh"
        h5 = hc[gen_name][0].Clone(gen_name+"_stack")
        stack.Add(h5)
        h5.SetLineColor(6)
        h5.SetLineWidth(2)
        if normalize:
            h5.Scale(1/h5.GetEntries())
        leg.AddEntry(h5, "gen p_{T}")

    leg.Draw("same");
    if option != "nostack":
        c.SaveAs(dir_name+"/"+h_name+"_stack.pdf")
    else:
        c.SaveAs(dir_name+"/"+h_name+"_nostack.pdf")
    return c, stack, h1, h2, h3, h4, leg
```

```python
no_cuts = False

c1 = draw_stack("genPtLow", "", rebinX = 2)
c2 = draw_stack("genPtHigh", "", rebinX = 2)
if no_cuts:
    c3 = draw_stack("MuonPtLow", "", rebinX = 2, title = "L2Muons")
    c4 = draw_stack("MuonPtHigh", "", rebinX = 2, title = "L2Muons")
    c5 = draw_stack("OfflinePtLow", "", rebinX = 2, title = "RSA")
    c6 = draw_stack("OfflinePtHigh", "", rebinX = 2, title = "RSA")
    c7 = draw_stack("SAPtLow", "", rebinX = 2, title = "SA")
    c8 = draw_stack("SAPtHigh", "", rebinX = 2, title = "SA")
else:
    c3 = draw_stack("MuonPtLow", "", rebinX = 2, title = "L2Muons with at least two DT or CSC chambers with hits")
    c4 = draw_stack("MuonPtHigh", "", rebinX = 2, title = "L2Muons with at least two DT or CSC chambers with hits")
    c5 = draw_stack("OfflinePtLow", "", rebinX = 2, title = "RSA muons with at least two DT or CSC chambers with valid hits")
    c6 = draw_stack("OfflinePtHigh", "", rebinX = 2, title = "RSA muons with at least two DT or CSC chambers with valid hits")
    c7 = draw_stack("SAPtLow", "", rebinX = 2, title = "SA muons with at least two DT or CSC chambers with valid hits")
    c8 = draw_stack("SAPtHigh", "", rebinX = 2, title = "SA muons with at least two DT or CSC chambers with valid hits")
canvas = TCanvas("", "", 800, 1600)
canvas.Divide(2,4)
canvas.cd(1)
c1[1].Draw()
c1[6].Draw("same")
canvas.cd(2)
c2[1].Draw()
c2[6].Draw("same")
canvas.cd(3)
c3[1].Draw()
c3[6].Draw("same")
canvas.cd(4)
c4[1].Draw()
c4[6].Draw("same")
canvas.cd(5)
c5[1].Draw()
c5[6].Draw("same")
canvas.cd(6)
c6[1].Draw()
c6[6].Draw("same")
canvas.cd(7)
c7[1].Draw()
c7[6].Draw("same")
canvas.cd(8)
c8[1].Draw()
c8[6].Draw("same")
canvas
```
