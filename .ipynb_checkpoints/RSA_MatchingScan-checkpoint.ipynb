```python
baseFilePath = "/Users/demattia/cmslpcbase/uscms_data/d3/demattia/DisplacedLeptons/RefittedStandAloneMuons/Analysis/CMSSW_5_3_8/src/AnalysisMacros/"
explicitPath = []
explicitPath.append("Original")
explicitPath.append("NoD0SigCut")
explicitPath.append("RemovedIsolation")
explicitPath.append("Eta2_RemovedAll_WiderDeltaR")
```

```python
class MatchedInfo:
    def __init__(self):
        self.matched = 0
        self.complementary = 0
    
def returnKey(name):
    if (name.find("ZZ") != -1 or name.find("DY") != -1 or name.find("QCD") != -1 or
       name.find("TT") != -1 or name.find("WZ") != -1 or name.find("WW") != -1 or name.find("WJets") != -1):
        return "Background"
    elif name.find("Data") != -1:
        return "Data"
    return name


matchedMap = {}
```

```python
def getMatchingEfficiency(inputFileName):
    matched = 0
    complementary = 0

    foundComplementary = False
    sampleName = ""
    inputFile = open(inputFileName)
    for line in inputFile:
        if line.find("processing") != -1:
            sampleName = line.split("/")[-1].split()[0]
            matchedInfo = MatchedInfo()
            if sampleName in matchedMap:
                matchedMap[sampleName].append(matchedInfo)
            else:
                matchedMap[sampleName] = [matchedInfo]
        if line.find("matched") != -1:
            # print line.strip("\n")
            matched = float(line.strip("\n").split("=")[1].strip())
            matchedMap[sampleName][-1].matched = matched
        elif line.find("complementary") != -1:
            # print line.strip("\n")
            complementary = float(line.strip("\n").split("=")[1].strip())
            matchedMap[sampleName][-1].complementary = complementary
            foundComplementary = True

        if foundComplementary:
            # print "match efficiency =", matched/(matched+complementary)
            foundComplementary = False
            sampleName = ""
```

```python
for name in explicitPath:
    getMatchingEfficiency(baseFilePath + name + "/log.txt")
```

```python
mergedMap = {}
for key in matchedMap:
    m_key = returnKey(key)
    if m_key not in mergedMap:
        mergedMap[m_key] = matchedMap[key]
    else:
        # print mergedMap[m_key]
        for a in range(len(mergedMap[m_key])):
            mergedMap[m_key][a].matched += matchedMap[key][a].matched
            mergedMap[m_key][a].complementary += matchedMap[key][a].complementary
```

```python
"""
Bar chart demo with pairs of bars grouped for easy comparison.
"""
import numpy as np
import matplotlib.pyplot as plt


n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()

plt.tight_layout()
plt.show()
```
