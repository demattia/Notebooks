```python
def draw_line(input_file_name, x_min, y_min, x_max, y_max, text_index = 4, text_size = 0.04):
    input_file = TFile(input_file_name, "R")
    c = input_file.FindObjectAny("c")
    text = c.GetListOfPrimitives()[text_index]
    text.SetText(text.GetX(), text.GetY(), "CMS Preliminary")
    text.SetTextSize(text_size)
    line = TLine(x_min, y_min, x_max, y_max)
    line.Draw()
    line.SetLineColor(8)
    line.SetLineWidth(3)
    name = input_file_name
    if name.endswith(".root"):
        name = name[:-5]
    c.SaveAs(name+"_expectedLimitLine.root")
    c.SaveAs(name+"_expectedLimitLine.pdf")
    c.SaveAs(name+"_expectedLimitLine.png")
    return c, line
```

```python
def shift_legend(name, c, ls_y1, ls_y2, y_min = 0.001):
    legend = c.GetListOfPrimitives()[13]
    legend.SetY1(legend.GetY1()+ls_y1)
    legend.SetY2(legend.GetY2()+ls_y2)
    if name.endswith(".root"):
        name = name[:-5]
    c.GetListOfPrimitives()[1].GetYaxis().SetRangeUser(y_min, 1000)
    c.SaveAs(name+"_expectedLimitLine.root")
    c.SaveAs(name+"_expectedLimitLine.pdf")
    c.SaveAs(name+"_expectedLimitLine.png")
    return c
```
