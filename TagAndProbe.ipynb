## Define RooFit variables and PDFs

```python
from ROOT import RooRealVar, RooFormulaVar, RooVoigtian, RooChebychev, RooArgList, RooArgSet, RooAddPdf, RooDataSet, RooCategory, RooSimultaneous
```

Define pdf for the passing probes

```python
mass = RooRealVar("mass", "mass", 60, 120)
# Construct signal pdf
mean = RooRealVar("mean", "mean", 90, 80, 100)
width = RooRealVar("width", "width", 2.4952, 1, 3)
width.setConstant(ROOT.kTRUE)
sigma = RooRealVar("sigma", "sigma", 1.2, 0.2, 2)
signalAll = RooVoigtian("signalAll", "signalAll", mass, mean, width, sigma)
# Construct background pdf
a0_all = RooRealVar("a0_all","a0_all",-0.1,-1,1)
a1_all = RooRealVar("a1_all","a1_all",0.004,-1,1)
backgroundAll = RooChebychev("backgroundAll","backgroundAll",mass,RooArgList(a0_all,a1_all))
# Construct composite pdf
sigAll = RooRealVar("sigAll", "sigAll", 2000, 0, 100000)
bkgAll = RooRealVar("bkgAll", "bkgAll", 100, 0, 1000)
modelAll = RooAddPdf("modelAll", "modelAll", RooArgList(signalAll, backgroundAll), RooArgList(sigAll, bkgAll))
```

Define pdf for all probes

```python
# Construct signal pdf. 
# NOTE that sigma is shared with the signal sample model
signalPass = RooVoigtian("signalPass","signalPass",mass,mean,width,sigma)
# Construct the background pdf
a0_pass = RooRealVar("a0_pass","a0_pass",-0.1,-1,1)
a1_pass = RooRealVar("a1_pass","a1_pass",0.5,-0.1,1)
backgroundPass = RooChebychev("backgroundPass","backgroundPass",mass,RooArgList(a0_pass,a1_pass))
# Construct the composite model
fraction = RooRealVar("fraction","fration",0.8,0.,1.)
sigPass = RooFormulaVar("sigPass", "@0*@1", RooArgList(sigAll, fraction))
bkgPass = RooRealVar("bkgPass", "bkgPass", 100, 0, 1000)
modelPass = RooAddPdf("modelPass", "modelPass", RooArgList(signalPass, backgroundPass), RooArgList(sigPass, bkgPass))
```

## Prepare datasets

Define binning

```python
import bisect

# ptBins = [23, 26, 30, 35, 40, 45, 50, 60, 70]
ptBins = [23, 100]

# This returns the position in the array
def findBins(pt1, pt2):
    return (bisect.bisect_right(ptBins, pt1)-1, bisect.bisect_right(ptBins, pt2)-1)

def findPosition(bin1, bin2):
    return bin1+len(ptBins)*bin2
```

```python
datasetAllMap = {}
datasetPassMap = {}
for ptBin1 in range(0, len(ptBins)):
    for ptBin2 in range(0, len(ptBins)):
        datasetAllMap[(ptBin1, ptBin2)] = RooDataSet("datasetAll_"+str(ptBin1)+"_"+str(ptBin2), "datasetAll_"+str(ptBin1)+"_"+str(ptBin2), RooArgSet(mass))
        datasetPassMap[(ptBin1, ptBin2)] = RooDataSet("datasetPass_"+str(ptBin1)+"_"+str(ptBin2), "datasetPass_"+str(ptBin1)+"_"+str(ptBin2), RooArgSet(mass))
```

### Event loop

Trigger efficiency for new trigger over old trigger

```python
oldTrigger = "HLT_L2DoubleMu23_NoVertex_v"
newTrigger = "HLT_L2DoubleMu23_NoVertex_2Cha_Angle2p5_v"
```

```python
hAllMap = {}
hPassMap = {}
for ptBin1 in range(0, len(ptBins)):
    for ptBin2 in range(0, len(ptBins)):
        hAllMap[(ptBin1,ptBin2)] = ROOT.TH1F("hAll"+"_"+str(ptBin1)+"_"+str(ptBin2), "All events passing old trigger "+str(ptBin1)+"_"+str(ptBin2), 100, 60, 120)
        hPassMap[(ptBin1,ptBin2)] = ROOT.TH1F("hPass"+"_"+str(ptBin1)+"_"+str(ptBin2), "All events passing old trigger and new trigger "+str(ptBin1)+"_"+str(ptBin2), 100, 60, 120)
        
        
hPt1 = ROOT.TH1F("hPt1", "muon1 p_T", 100, 0, 100)
hPt2 = ROOT.TH1F("hPt2", "muon2 p_T", 100, 0, 100)
```

```python
def passSelection(track):
    if ( track.pt > 26 and abs(track.eta) < 2 and track.isolation/track.pt < 0.1 and track.trackerLayersWithMeasurement >= 6
         and track.trackQuality and track.dxy < 30. and track.dz < 30. ):
        return True

def passDileptonSelection(track1, track2, cosine):
    if track1.charge == track2.charge: return False
    if utils.deltaR(track1.phi, track1.eta, track2.phi, track2.eta) < 0.2: return False
    # cosineAndMass = computeCosineAndMass(track1, track2)
    if cosine <= -0.79: return False
    return True

def computeCosineAndMass(mu1, mu2):
    muon1 = ROOT.TLorentzVector()
    muon2 = ROOT.TLorentzVector()
    muon1.SetPtEtaPhiM(mu1.pt, mu1.eta, mu1.phi, muMass)
    muon2.SetPtEtaPhiM(mu2.pt, mu2.eta, mu2.phi, muMass)
    cosine = math.cos(muon1.Angle(muon2.Vect()))
    mass = (muon1+muon2).M()
    return (cosine, mass)

def fillTriggerMatchedTrack(track, triggerObjects, matchedTracks):
    for triggerMuon in triggerObjects:
        if utils.deltaR(triggerMuon.phi, triggerMuon.eta, track.phi, track.eta) < triggerMatchDeltaR and passSelection(track):
            matchedTracks.append(track)

def fillSingleCandidate(track1, track2, histoMap, datasetMap):
    cosineAndMass = computeCosineAndMass(track1, track2)
    if not passDileptonSelection(track1, track2, cosineAndMass[0]): return False
    mass.setVal(cosineAndMass[1])
    histoMap[findBins(track1.pt, track2.pt)].Fill(mass.getVal())
    datasetMap[findBins(track1.pt, track2.pt)].add(RooArgSet(mass))
    return True

def fillCandidates(matchedTracks, histoMap, datasetMap):
    if len(matchedTracks) == 2:
        fillSingleCandidate(matchedTracks[0], matchedTracks[1], histoMap, datasetMap)
    elif len(matchedTracks) > 2:
        for i in range(0, len(matchedTracks)):
            for j in range(i+1, len(matchedTracks)):
                fillSingleCandidate(matchedTracks[i], matchedTracks[j], histoMap, datasetMap)
```

```python
allCandidates = 0
passCandidates = 0

processedEvents = 0

for event in tree:
    # if processedEvents > 10000:
    #     break

    # Get the trigger outcome for this event
    oldTriggerFired = False
    newTriggerFired = False
    for name in event.triggerNames:
        if name.find(oldTrigger) != -1: oldTriggerFired = True
        if name.find(newTrigger) != -1: newTriggerFired = True
    
    # If none of the two triggers fired we can skip the event
    if not oldTriggerFired and not newTriggerFired:
        continue

    # print "event number", processedEvents
    processedEvents += 1

    # print "run, event =", str(event.run)+",", event.event
    oldTriggerObjects = event.triggerFilterObjectsMap["hltL2DoubleMu23NoVertexL2PreFiltered"]
    newTriggerObjects = event.triggerFilterObjectsMap["hltL2DoubleMu23NoVertexL2Filtered2ChaAngle2p5"]
    
    matchedTracksOldTrigger = []
    matchedTracksNewTrigger = []
    for track in event.tracks:
        # Find a matching trigger object in DeltaR
        fillTriggerMatchedTrack(track, oldTriggerObjects, matchedTracksOldTrigger)
        fillTriggerMatchedTrack(track, newTriggerObjects, matchedTracksNewTrigger)
    
    if len(matchedTracksOldTrigger) > 2:
        # print "too many matches old trigger:", len(matchedTracksOldTrigger)
        pass
    if len(matchedTracksNewTrigger) > 2:
        # print "too many matches new trigger:", len(matchedTracksNewTrigger)
        pass
    
    if oldTriggerFired:
        if len(matchedTracksOldTrigger) == 2:
            # if matchedTracksOldTrigger[1].pt < matchedTracksOldTrigger[0].pt:
            #     print matchedTracksOldTrigger[0].pt
            hPt1.Fill(matchedTracksOldTrigger[0].pt)
            hPt2.Fill(matchedTracksOldTrigger[1].pt)
        # print "old trigger fired"
        fillCandidates(matchedTracksOldTrigger, hAllMap, datasetAllMap)

    # Note: we require both triggers. This is only needed in data because the old trigger is prescaled. In MC the old trigger always fires if the new one fires.
    if oldTriggerFired and newTriggerFired:
        # print "new trigger fired"
        fillCandidates(matchedTracksNewTrigger, hPassMap, datasetPassMap)

print "all candidates =", hAllMap[0,0].GetEntries()
print "pass candidates =", hPassMap[0,0].GetEntries()
```
