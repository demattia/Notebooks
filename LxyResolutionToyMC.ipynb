```python
import ROOT
import rootprint
import rootnotes
import math
```

```python
hGausX = ROOT.TH1F("gausX", "gausX", 100, -10, 10)
hGausY = ROOT.TH1F("gausY", "gausY", 100, -10, 10)
hLxy = ROOT.TH1F("Lxy", "Lxy", 100, 0, 10)
for i in range(100000):
    svx = ROOT.gRandom.Gaus(0,1)
    svy = ROOT.gRandom.Gaus(0,1)
    hGausX.Fill(svx)
    hGausY.Fill(svy)
    hLxy.Fill(math.sqrt(svx**2+svy**2))
```

```python
c = ROOT.TCanvas()
c.Divide(2,2)
c.cd(1)
hGausX.Draw()
c.cd(2)
hGausY.Draw()
f = hGausY.Fit("gaus")
c.cd(3)
hLxy.Draw()
c
```
