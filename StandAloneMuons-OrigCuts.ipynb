```python
import ROOT
import rootnotes
import rootprint
import math
import rootpy
from rootpy.io import root_open
```

```python
rootFile = root_open("/Users/demattia/cmslpcbase/uscms_data/d3/demattia/DisplacedLeptons/"+
                     "RefittedStandAloneMuons/Analysis/CMSSW_5_3_8/src/AnalysisMacros/DileptonMacros/output/"+
                     "output_2saMu_rereco.root")
```

```python
def excluded(histo):
    if histo.find("nHitsBeforeVertex") != -1: return True
    if histo.find("minMissingHitsAfterVertex") != -1: return True
    if histo.find("nMissingHitsAfterVertex") != -1: return True
    if histo.find("maxMissingHitsAfterVertex") != -1: return True
```

```python
nMinus1 = {}

# %%rootprint
for path, dirs, objects in rootFile.walk():
    for histo in objects:
        print histo
        if histo.find("nMinus1") != -1:
            if excluded(histo): continue
            histoType = histo.split("_")[-1]
            if histoType in nMinus1:
                nMinus1[histoType].append(histo)
            else:
                nMinus1[histoType] = [histo]
        else:
            # print histo
            pass
```

```python
nMinus1
```

```python
def drawPlots(newCanvas, histoType):
    index = 1
    pad = ROOT.TCanvas()
    for histo in nMinus1[histoType]:
        print histo
        pad = getattr(rootFile, histo)
        newCanvas.cd(index)
        pad.DrawClonePad()
        index += 1
    return index
```

### Control Collinearity Cuts (no lifetime cuts)

```python
newCanvas = ROOT.TCanvas("newCanvas", "newCanvas", 1000, 1000)
newCanvas.Divide(2,3)
index = drawPlots(newCanvas, "controlCollCuts")
pad = rootFile.minLeptonAbsD0Sig_bestCand_deltaPhiG90_removedLifetimeCollCuts
newCanvas.cd(index)
pad.DrawClonePad()
newCanvas
```

### Loose Control Collinearity Cuts

```python
newCanvas = ROOT.TCanvas("newCanvas", "newCanvas", 1000, 1000)
newCanvas.Divide(2,3)
drawPlots(newCanvas, "looseControlCollCuts")
newCanvas
```
